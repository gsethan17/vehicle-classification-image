# mat2csv

```python
import scipy.io
import pandas as pd
import numpy as np


def mat2csv(matfile):
    imgs = []
    anno1 = []
    anno2 = []
    anno3 = []
    anno4 = []
    anno5 = []
    anno6 = []
    classes = []
    datas = scipy.io.loadmat(matfile +'.mat')
    key_list = list(datas.keys())

    for key in key_list:
        if key == 'annotations' or key == 'class_names':
            data = np.array(datas[key])
            data = data.reshape((data.shape[1],))

            for comp in data:
                if key == 'annotations':
                    imgs.append(comp[0][0])
                    anno1.append(comp[1][0][0])
                    anno2.append(comp[2][0][0])
                    anno3.append(comp[3][0][0])
                    anno4.append(comp[4][0][0])
                    anno5.append(comp[5][0][0])
                    anno6.append(comp[6][0][0])
                elif key == 'class_names':
                    classes.append(comp[0])

    print(data)
    
    # 데이터프레임으로 변환
    dic = {
        'img':imgs,
        'anno1':anno1,
        'anno2':anno2,
        'anno3':anno3,
        'anno4':anno4,
        'anno5':anno5,
        'anno6':anno6,
    }
    df = pd.DataFrame(dic)
    df.to_csv(matfile+'.csv', index=False)

    dic = {
        'class':classes,
    }
    df = pd.DataFrame(dic)
    df.to_csv(matfile+'_class.csv')
    # df.to_csv(matfile +'.csv', index=False)
```

```python
mat_name = ['cars_annos']
```

```python
for file in mat_name:
    annos = mat2csv(file)

# 3m 52.6s 소요
```
